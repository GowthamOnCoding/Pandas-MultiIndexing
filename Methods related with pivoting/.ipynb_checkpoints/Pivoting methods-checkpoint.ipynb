# Pivoting methods

```python
import pandas as pd
sale=pd.read_csv("salesmen.csv")
sale["Salesman"].astype("category")
sale.head()
```

```python
sale.pivot(index="Date",columns="Salesman",values="Revenue")
```

```python
food=pd.read_csv("foods.csv")
food.head()
```

```python
food.pivot_table(index="Gender",values="Spend",aggfunc=sum)
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc=sum)
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc=sum,columns=["City"])
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc=sum,columns=["Frequency","City"])
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc="count",columns=["City"])
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc="min",columns=["City"])
```

```python
food.pivot_table(index=["Gender","Item"],values="Spend",aggfunc="max",columns=["City"])
```

```python
quarter=pd.read_csv("quarters.csv")
quarter.head(3)
```

```python
pd.melt(id_vars="Salesman",frame=quarter)
```

```python
pd.melt(id_vars="Salesman",frame=quarter,var_name="Quarter",value_name="Revenue")
```
